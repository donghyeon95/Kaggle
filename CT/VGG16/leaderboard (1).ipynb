```python
def majority_vote(li):
    submit = pd.read_csv('baseline.csv')
    result = []
    for i in range(len(submit)):
        score = 0
        for j in range(len(li)):
            score += li[j][i]
        if round(len(li)/2+0.51) <= score:
            result.append(1)
        else:
            result.append(0)


    submit["COVID"] = result  
    return submit
```

```python
def powerset(s):
    masks = [ 1 << i for i in range(len(s)) ]
    for i in range( 1 << len(s) ):
        yield [ss for ss,mask in zip(s,masks) if mask & i ]        
    
score_dict = {}
predict_model = LinearRegression().fit(X, y)

for power in powerset([x for x in range(len(li))]):
    if len(power) > 0:
        
        models = []
        for i in power:
            models.append(li[i])
            
        result =  majority_vote(models)
        # print(models)
        score_dict[str(power)] = predict_model.predict(pd.DataFrame(result['COVID']).T)

test = sorted(score_dict.items(), key=lambda x: x[1], reverse=True)
test
```
